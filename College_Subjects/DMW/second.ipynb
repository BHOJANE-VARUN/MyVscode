```python
# df.iloc[:2, 0]
# one based indexing
def coovariane(i,j):
    firstMean = df.iloc[:,i].mean()
    secondMean = df.iloc[:,j].mean()
    sum =0
    for x in range(0,df.iloc[:,i].size):
        sum += df.iloc[x,i]-firstMean*df.iloc[x,j]-secondMean
    return sum/df.iloc[:,i].size
```

```python
import pandas as pd
X = pd.read_csv('input_data.csv')
y = pd.read_csv('output_data.csv')
y.head(10)
```

```python
import sklearn.model_selection as ms;
x_train,x_test,y_train,y_test = ms.train_test_split(X,y,test_size=0.2,random_state=40)
```

```python
from sklearn.linear_model import SGDRegressor
sgd_model = SGDRegressor(max_iter=1000, tol=1e-3)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(x_train)
X_test = scaler.transform(x_test)
```

```python
sgd_model.fit(X_train,y_train)
```

```python
X.head(20)
```

```python

ls = [[25200,23],[26788,33]]
mytest = pd.DataFrame(ls)
mytest = scaler.transform(mytest)
sgd_model.predict(mytest)
```

```python
sgd_model.score(X_test,y_train)
```

```python
sgd_model.score(X_test,y_test)*100
```

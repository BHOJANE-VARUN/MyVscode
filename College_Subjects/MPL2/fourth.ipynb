```python
import pandas as pd
df1 = pd.read_csv('input.csv')
df1
```

```python
df2 = pd.read_csv('newinput.csv')
df2
```

```python
df2['nextstationtimining'] = df2['nextstationtime']
df2 = df2.drop(['nextstationtime'],axis='columns')
df2
```

```python
X = pd.concat([df1,df2])
X
```

```python
def convert_time_to_seconds(time_str):
  """
  Converts a string of the format "hrs-min-sec" into seconds as an integer.

  Args:
    time_str: A string of the format "hrs-min-sec".

  Returns:
    An integer representing the time in seconds.
  """

  hours, minutes, seconds = time_str.split(':')
  if hours=='':
    hours =0
  if minutes=='':
    minutes =0
  if seconds=='':
    seconds=0
  hours = int(hours)
  minutes = int(minutes)
  seconds = int(seconds)
  return hours * 3600 + minutes * 60 + seconds

# Example usage
time_str = "2:30:45"
seconds = convert_time_to_seconds(time_str)
print(f"{time_str} in seconds is {seconds}")
```

```python
X['time'] = X['time'].apply(convert_time_to_seconds)
X['nextstationtimining'] = X['nextstationtimining'].apply(convert_time_to_seconds)
X
```

```python
inpu = X.drop(['bus_id','nextstationtimining'],axis='columns')
inpu
```

```python
out = X['nextstationtimining']
out
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(inpu,out)
```

```python
model.score(inpu,out)
```

```python
model.predict([[10000,1,1,0,0]])
```

```python
import pickle
with open('bus_model','wb') as f:
    pickle.dump(model,f)
```

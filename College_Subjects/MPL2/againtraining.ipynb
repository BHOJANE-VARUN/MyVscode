```python
import pandas as pd
df = pd.read_csv('newinput.csv')
df
```

```python
def convert_time_to_seconds(time_str):
  """
  Converts a string of the format "hrs-min-sec" into seconds as an integer.

  Args:
    time_str: A string of the format "hrs-min-sec".

  Returns:
    An integer representing the time in seconds.
  """

  hours, minutes, seconds = time_str.split(':')
  if hours=='':
    hours =0
  if minutes=='':
    minutes =0
  if seconds=='':
    seconds=0
  hours = int(hours)
  minutes = int(minutes)
  seconds = int(seconds)
  return hours * 3600 + minutes * 60 + seconds
```

```python
df['time'] = df['time'].apply(convert_time_to_seconds)
df['nextstationtime'] = df['nextstationtime'].apply(convert_time_to_seconds)
df
```

```python
X = df.drop(['bus_id','nextstationtime'],axis='columns')
X
```

```python
y = df['nextstationtime']
y
```

```python
import pickle
with open('bus_model', 'rb') as file:
        model = pickle.load(file)
model
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```

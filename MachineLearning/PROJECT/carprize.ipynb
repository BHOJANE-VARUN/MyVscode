```python
import pandas as pd
df1 = pd.read_csv('UsedCars_Pune.csv')
df1['city'].describe()
```

```python
df1 = df1.drop(['Unnamed: 0','brand','city'],axis='columns')
df1
```

```python
# 0 for petrol and 1 for diesel
# df1.loc[df1['fuel_type'] != 'Petrol', 'fuel_type'] = 0
# df1.loc[df1['fuel_type'] == 'Petrol', 'fuel_type'] = 1

# df1['petrol'] = df1['fuel_type'].apply(lambda x: 1 if x==1 else 0)
# df1['diesel'] = df1['fuel_type']
df1.drop(['fuel_type'],axis='columns',inplace=True)
df1
```

```python
df1['car_name'] = df1['full_model_name'] + df1['model_name']
df1
```

```python
df1 = df1.drop(['full_model_name','model_name'],axis='columns')
df1
```

```python
lessdes = df1.groupby('car_name')['car_name'].agg('count')
lessdes
```

```python
one = lessdes[lessdes==4]
one.describe()
```

```python
df2 = df1
```

```python
df2.car_name = df2.car_name.apply(lambda x: "other" if x in one else x)
df2
```

```python
df3 = df2[df2['car_name'] != 'other']
```

```python
df3
```

```python
dump = pd.get_dummies(df3.car_name).drop(['Audi A6 2.0 TDI PremiumA6'],axis='columns')
```

```python
df4 = pd.concat([df3,dump],axis='columns')
df5 = df4.drop(['car_name'],axis='columns')
df5
```

```python
X = df5.drop(['price'],axis='columns')
X.head()
```

```python
y = df5['price']
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

cross_val_score(LinearRegression(), X, y, cv=cv)
```

```python
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X,y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                'positive': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(X,y)
```

```python
model = Lasso(alpha=1,selection='cyclic')
```

```python
model.fit(X,y)
```

```python
model.score(X_test,y_test)
```

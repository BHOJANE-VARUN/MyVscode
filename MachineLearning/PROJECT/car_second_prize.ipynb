```python
import pandas as pd
df1 = pd.read_csv('autos.csv')
df1['name'].describe()
```

```python
lessdes = (df1.groupby('name')['name'].agg('count'))
lessdes
```

```python
ls = list(lessdes[lessdes>412].keys())
ls
```

```python
df2 = df1[df1['name'].isin(ls)]
df2
```

```python
df2 = df2.drop(['seller','offerType','dateCrawled','dateCreated','nrOfPictures','lastSeen'],axis='columns')
df2
```

```python
df2.isnull().sum()
```

```python
df3 = df2.dropna()
```

```python
df4 = df3.drop(['model','brand'],axis='columns')
df4
```

```python
standder = df4['price'].std()
meanofdf = df4['price'].mean()
# prics = (df4.groupby('price')['price'].agg('count'))
# prics[prics<2].count()
df5 = df4[(df4['price']>(standder-meanofdf)) & (df4['price']<(standder+meanofdf))]
```

```python
df5['price'].describe()
```

```python
import matplotlib.pyplot as plt
temp = df5[df5['name']=='BMW_316i']
temp
plt.scatter(temp['yearOfRegistration'],temp['price'])
```

```python
df6 = df5[(df5['yearOfRegistration']>1992) & (df5['yearOfRegistration']<2005)]
df6
```

```python
temp2 = df6[df6['name']=='BMW_316i']
temp2
plt.scatter(temp2['yearOfRegistration'],temp2['price'])
```

```python
df7 = df6.drop(['postalCode'],axis='columns')
df7
```

```python
df8 = pd.concat([pd.get_dummies(df7['notRepairedDamage']),df7.drop(['notRepairedDamage'],axis='columns')],axis='columns')
```

```python
df8.drop(['index'],axis='columns',inplace=True)
```

```python
df9 = pd.concat([pd.get_dummies(df8['fuelType']),df8.drop(['fuelType'],axis='columns')],axis='columns')
df9
```

```python
df10 = pd.concat([pd.get_dummies(df9['gearbox']),df9.drop(['gearbox'],axis='columns')],axis='columns')
df10
```

```python
df11 = pd.concat([pd.get_dummies(df10['vehicleType']),df10.drop(['vehicleType'],axis='columns')],axis='columns')
df11 = pd.concat([pd.get_dummies(df11['abtest']),df11.drop(['abtest'],axis='columns')],axis='columns')
df11 = pd.concat([pd.get_dummies(df11['name']),df11.drop(['name'],axis='columns')],axis='columns')
df11
```

```python
Y = df11['price']
Y
```

```python
X = df11.drop(['price'],axis='columns')
X
```

```python
X.shape
```

```python
Y.shape
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

cross_val_score(LinearRegression(), X, Y, cv=cv)
```

```python
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_best_model_using_gridsearchcv(X,y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion' : ['mse','friedman_mse'],
                'splitter': ['best','random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs =  GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

find_best_model_using_gridsearchcv(X,Y)
```

```python
x_trai,x_tes,y_trai,y_tes = train_test_split(X,Y,test_size=0.2,random_state=42)
```

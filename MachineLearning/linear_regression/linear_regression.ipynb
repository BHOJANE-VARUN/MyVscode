```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataframe = pd.read_csv('Salary_dataset.csv',usecols=['YearsExperience','Salary'])
print(dataframe.describe())
```

```python
X = dataframe.drop(columns=['YearsExperience'])
Y = dataframe['YearsExperience']

# for x in range(0,25):
#     Yearexp.append(dataframe.YearsExperience.loc[x])
#     Salary.append(dataframe.Salary.loc[x])
# siz = np.array(Yearexp)
# weig = np.array(Salary)
# plt.ylabel('experience in years')
# plt.xlabel('Salary')

plt.plot(X,Y,'o')
```

```python
import sklearn.linear_model as lm;
import sklearn.model_selection as ms;
```

```python
model = lm.LinearRegression()
x_train,x_test,y_train,y_test = ms.train_test_split(X,Y,test_size=0.2,random_state=40)
print(x_train)
# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(x_train)
X_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import SGDRegressor
sgd_model = SGDRegressor(max_iter=1000, tol=1e-3)
```

```python
sgd_model.fit(X_train,y_train)
```

```python
plt.scatter(X_train,y_train)
plt.plot(X_train,sgd_model.intercept_[0] + sgd_model.coef_[0]*X_train, '-r')
sgdpred = sgd_model.predict(X_test)
plt.scatter(X_test,sgdpred)
# mytest = [[9.7]]
# test = scaler.fit_transform(mytest)
# pred = sgd_model.predict(test)
# print(pred)
# plt.scatter(X,Y)
# plt.scatter(x_test,y_test)
```

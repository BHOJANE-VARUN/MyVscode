```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataframe = pd.read_csv('Salary_dataset.csv')
print(dataframe)
```

```python
Yearexp = []
Salary = []

for x in range(0,25):
    Yearexp.append(dataframe.YearsExperience.loc[x])
    Salary.append(dataframe.Salary.loc[x])
siz = np.array(Yearexp)
weig = np.array(Salary)
plt.ylabel('experience in years')
plt.xlabel('Salary')

plt.plot(siz,weig,'o')
```

```python
import sklearn.linear_model as lm;
```

```python
model = lm.LinearRegression();
model.fit(siz,weig)
```

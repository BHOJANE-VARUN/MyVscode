```python
import pandas as pd
cars = pd.read_csv('carprices.csv')
cars
```

```python
dum = pd.get_dummies(cars['Car Model'])
```

```python
df = pd.concat([cars,dum],axis='columns')
df
```

```python
X = df.drop(['Car Model','Audi A5','Sell Price($)'],axis='columns')
X
```

```python
y = df['Sell Price($)']
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.predict([[45000,4,0,1]])
# Mercedes = array([[36991.31721061]])
# BMW = array([[11450.86522658]])
# Accuracy = 0.9417050937281082
```

```python
model.predict([[86000,7,1,0]])
```

```python
model.score(X,y)
```

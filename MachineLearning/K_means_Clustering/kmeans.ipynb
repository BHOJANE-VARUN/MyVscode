```python
import pandas as pd
df = pd.read_csv('data1.txt')
df
```

```python
from matplotlib import pyplot as plt
plt.scatter(df['0.89'],df['6.35'])
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3)
```

```python
pred = model.fit_predict(df[['6.35','0.89']])
df['pred'] = pred
df
```

```python
df[df['pred']==0]
plt.scatter(df[df['pred']==0]['0.89'],df[df['pred']==0]['6.35'],color = 'green')
plt.scatter(df[df['pred']==1]['0.89'],df[df['pred']==1]['6.35'],color = 'red')
plt.scatter(df[df['pred']==2]['0.89'],df[df['pred']==2]['6.35'],color = 'blue')
plt.scatter(model.cluster_centers_[0][0],model.cluster_centers_[0][1],color='yellow',marker='*')
plt.scatter(model.cluster_centers_[1][0],model.cluster_centers_[1][1],color='yellow',marker='*')
plt.scatter(model.cluster_centers_[2][0],model.cluster_centers_[2][1],color='yellow',marker='*')
model.cluster_centers_[0][1]
```

```python
ls = []
for k in range(1,20):
    km = KMeans(n_clusters=k)
    km.fit(df[['6.35','0.89']])
    ls.append(km.inertia_)
```

```python
plt.plot(ls)
```
